```python
modified_data = []

# Iterate through each list in all_data
for data_list in all_data:
    # Initialize a new list to store the modified data for the current list
    modified_list = []
    
    # Iterate through each data point in the current list
    for data_point in data_list:
        # Convert the data point to string based on its value
        if data_point >=2:
            modified_list.append('>=2')
        elif data_point < 2:
            modified_list.append('<2')
        
    
    # Append the modified list to modified_data
    modified_data.append(modified_list)



# Count the occurrences of each unique value in modified_data[0]
value_counts = Counter(modified_data[3])

# Print the counts of each unique value
print(value_counts)
modified_test=[]
for i in range(len(test_data)):
    # Convert the data point to string based on its value
    if test_data[i] >= 2:
        modified_test.append('>=2')
    elif test_data[i] < 2:
        modified_test.append('<2')
value_counts = Counter(modified_test)

# Print the counts of each unique value
print(value_counts)

```

now making a markov of order 5 to predict

```python
transition_counts = defaultdict(lambda: defaultdict(int))

# Iterate through each list in modified_data
from collections import defaultdict

# Initialize a defaultdict to store transition counts
transition_counts = defaultdict(lambda: defaultdict(int))

# Iterate through each list in modified_data
for data_list in modified_data:
    # Iterate through each combination of five consecutive data points
    for i in range(len(data_list) - 5):
        # Get the current five data points
        current_points = tuple(data_list[i:i+5])
        # Update the transition count
        transition_counts[current_points][data_list[i+5]] += 1

# Print the transition counts
for current_points, transitions in transition_counts.items():
    print(f"Current Points: {current_points}")
    for next_state, count in transitions.items():
        print(f"Next State: {next_state}, Count: {count}")
    print()
transition_counts
```

now a transitional probability matrix

```python
transition_probs = {}

# Iterate through each combination of five consecutive data points
for current_points, transitions in transition_counts.items():
    # Calculate the total count of transitions from the current state
    total_count = sum(transitions.values())
    # Initialize a dictionary to store transitional probabilities for the current state
    probs = {}
    # Calculate the probability of transitioning to each next state
    for next_state, count in transitions.items():
        probs[next_state] = count / total_count
    # Store the transitional probabilities for the current state
    transition_probs[current_points] = probs

# Print the transitional probabilities
for current_points, probs in transition_probs.items():
    print(f"Current Points: {current_points}")
    for next_state, prob in probs.items():
        print(f"Next State: {next_state}, Probability: {prob}")
    print()
```

now the predictions

```python
# Initialize counters for correct predictions and total predictions
# Initialize counters for correct predictions and total predictions
correct_predictions = 0
total_predictions = 0

# Iterate through each combination of five consecutive data points in test_data
for i in range(len(modified_test) - 5):
    # Get the current five data points
    current_points = tuple(modified_test[i:i+5])
    # Check if we have transitional probabilities for the current state
    if current_points in transition_probs:
        # Predict the next state using transitional probabilities
        next_state_probs = transition_probs[current_points]
        predicted_next_state = max(next_state_probs, key=next_state_probs.get)
        # Get the actual next state
        actual_next_state = modified_test[i+5]
        # Compare the predicted next state with the actual next state
        if predicted_next_state == actual_next_state:
            correct_predictions += 1
        total_predictions += 1

# Calculate accuracy and precision
accuracy = correct_predictions / total_predictions if total_predictions > 0 else 0


# Print the results
print(f"Correct Predictions: {correct_predictions}")
print(f"Total Predictions: {total_predictions}")
print(f"Accuracy: {accuracy}")


```
